## Curves to graph

$y = 0x + 1$ and label the curve "O(1)"

$y = log(x)$ and label the curve "O(log n)"

$y = x$ and label the curve "O(n)"

$y = x * log(x)$ and label it "O(n log n)"

$y = x^2$ and label it "O(n^2)"

$y = 2^x$ and label it "O(2^n)"

$y = n!$ and label is "O(n!)"

$y = n^n$ and label it "O(n^n)"

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
x = range(1,10)
y_01 = [(0 * n) + 1 for n in x]
y_log = [math.log(n) for n in x]
y_n = [n for n in x]
y_logn = [(n * math.log(n)) for n in x]
y_2n = [n ** 2 for n in x]
y_n2 = [2 ** n for n in x]
y_nf = [(math.factorial(n)) for n in x]
y_nn = [n ** n for n in x]

plt.figure(figsize=(16,6))

plt.plot(x, y_01, label = "$O(1)$")
plt.plot(x, y_log, label = "$O(log n)$")
plt.plot(x, y_n, label = "$O(n)$")
plt.plot(x, y_logn, label = "$O(n log n)$")
plt.plot(x, y_2n, label = "$O(n^2)$")
plt.plot(x, y_n2, label = "$O(2^n)$")
plt.plot(x, y_nf, label = "$O(n!)$")
plt.plot(x, y_nn, label = "$O(n^n)$")

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.yscale("log")

plt.show()

```

```python
n_rows = 2
n_cols = 4
x = range(1,100)
y_01 = [(0 * n) + 1 for n in x]
y_log = [math.log(n) for n in x]
y_n = [n for n in x]
y_logn = [(n * math.log(n)) for n in x]
y_2n = [n ** 2 for n in x]
y_n2 = [2 ** n for n in x]
y_nf = [(math.factorial(n)) for n in x]
y_nn = [n ** n for n in x]

plt.figure(figsize=(16,6))

plt.subplot(n_rows, n_cols, 1)
plt.plot(x, y_01, label = "$O(1)$")
plt.title("$O(1)$")

plt.subplot(n_rows, n_cols, 2)
plt.plot(x, y_log, label = "$O(log n)$")
plt.title("$O(log n)$")

plt.subplot(n_rows, n_cols, 3)
plt.plot(x, y_n, label = "$O(n)$")
plt.title("$O(n)$")

plt.subplot(n_rows, n_cols, 4)
plt.plot(x, y_logn, label = "$O(n log n)$")
plt.title("$O(n log n)$")

plt.subplot(n_rows, n_cols, 5)
plt.plot(x, y_2n, label = "$O(n^2)$")
plt.title("$O(n^2)$")

plt.subplot(n_rows, n_cols, 6)
plt.plot(x, y_n2, label = "$O(2^n)$")
plt.title("$O(2^n)$")

plt.subplot(n_rows, n_cols, 7)
plt.plot(x, y_nf, label = "$O(n!)$")
plt.title("$O(n!)$")

plt.subplot(n_rows, n_cols, 8)
plt.plot(x, y_nn, label = "$O(n^n)$")
plt.title("$O(n^n)$")

plt.suptitle('Big O Notation')
# plt.xlabel('Elements')
# plt.ylabel('Operations')


plt.show()
```
