Q1:Use matplotlib to plot the following equation:

y = x^2 - x + 2 

You'll need to write the code that generates the x and y points and add an anotation for the point ```0,0``` the origin

````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
````

````python
x = np.array(range(100))
y = (x ** 2) - (x+2)

plt.plot(x,y)

plt.text(0,0, '(0,0)', fontsize=10)
plt.show()
````

````python
x = range(5)
y = [(n ** 2) - (n + 2) for n in x]
plt.plot(x,y)
plt.text(0,0, '(0,0)', fontsize=10)
plt.show()
````

Q4 Combine the figures you created in the last step into one large figure with 4 subplots

````python
n_rows = 2
n_cols = 2

plt.subplot(n_rows, n_cols, 1)
x = np.array(range(5))
y = [n ** x for n in x]
plt.title("Plot of y = 2^n ")
plt.plot(x,y)
plt.show()

plt.subplot(n_rows, n_cols, 2)
x = list(range(-4,4))
y = [math.tan(n) for n in x]
plt.title("Plot of y = tan(x)")
plt.plot(x,y)
plt.show()

plt.subplot(n_rows, n_cols, 3)
x = [1,5,10,15]
y = [n ** 3 for n in x]
plt.title("Plot looking at y = n ** 3")
plt.plot(x,y)
plt.show()

plt.subplot(n_rows, n_cols, 4)
x = np.array(range(50))
y = [math.sqrt(n) for n in x]

plt.title('Plot of y = \sqrt{x}')
plt.plot(x,y)
plt.show()
````

Q4 Comine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

````python
x = np.array(range(5))
y1 = [2 ** n for n in x]
plt.figure(figsize=(14,6))
y2 = [math.tan(n) for n in x]
y3 = [n ** 3 for n in x]
y4 = [math.sqrt(n) for n in x]

plt.plot(x,y1, label = "Plot of y = 2^n ")
plt.plot(x,y2, label= "Plot of y = tan(x)")
plt.plot(x, y3, label = "Plot looking at y = n ** 3" )
plt.plot(x, y4, label = 'Plot of y = $\sqrt{x}$')

plt.legend()
plt.title("Combined graph for matplot exercises")

plt.show()
````
